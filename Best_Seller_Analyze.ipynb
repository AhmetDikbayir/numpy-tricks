## Bütün veride tekrarlayan satır varsa onu veren kod

```python
df.duplicated().sum()
```

## Tekrarlayan satırlardaki verileri görmek için yazılması gereken kod

```python
df.loc[df.duplicated(keep = 'False'), :]
###keep = False demek tekrarlayan bütün satırları getirmesini söylemek demek
df.loc[df.duplicated(keep = 'first'), :]
###keep = 'First' dersek ilk tekrarlayan satırı getirir. ikinci satırı getirmez.
df.loc[df.duplicated(keep = 'last'), :]
###Tekrarlayan satırların ikincisini getirir. İlk satırı getirmez.
```

### Tekrarlayan satırları silmek için olan alternatif kodlar

```python
df.drop_duplicates(keep= 'first')
# ilk tekrarlayan verileri tutar.
df.drop_duplicates(keep = 'last')
# ikinci tekrarlayan verileri tutar diğerlerini siler.
df.drop_duplicates(keep = False)
# Tekrarlayan bütün verileri siler.
```

### Birden fazla sütundaki tekrarlayan verileri silmek istiyorsak kullanacağımız kod

```python
df.drop_duplicates(subset=['Author','Name'])
```
